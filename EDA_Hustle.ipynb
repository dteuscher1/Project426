```python
import pandas as pd
from plotnine import ggplot, geom_line, aes, labs, geom_boxplot, geom_histogram, ggtitle
```

```python
data = pd.read_csv('model_data.csv')
#data['WINNER_cat'] = data['WINNER'].astype('category')
```

```python
#  All of the below histograms show how the differences should be (and are) centered at 0 and perfectly symmetrical, so feel free to discard any you don't want

ggplot(data, aes(x = 'DEFLECTIONS_DIFF')) + geom_histogram(binwidth = 1) + ggtitle("Histogram of DEFLECTIONS_DIFF")
```

```python
ggplot(data, aes(x = 'DEF_BOXOUT_DIFF')) + geom_histogram(binwidth = 1) + ggtitle("DEF_BOXOUT_DIFF")
```

```python
ggplot(data, aes(x = 'OFF_BOXOUT_DIFF')) + geom_histogram(binwidth = 1) + ggtitle("Histogram of OFF_BOXOUT_DIFF")
```
